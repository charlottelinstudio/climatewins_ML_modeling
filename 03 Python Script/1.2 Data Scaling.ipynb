# 3. Data Scaling

```python
# Initiate the scaler
scaler = StandardScaler()
```

```python
# Identify numeric columns (excluding identifiers)
exclude_cols = ['DATE', 'id']
num_cols = df.drop(columns=[col for col in exclude_cols if col in df.columns]).select_dtypes(include=['float64', 'int64']).columns
```

```python
# Fit and transform the numeric columns
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
# Verify mean and std
print("Mean of scaled features (should be ~0):")
print(df[num_cols].mean().round(3).head())
print("\nStandard deviation of scaled features (should be ~1):")
print(df[num_cols].std().round(3).head())
```
